### Lybraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x= [0,1,2,3,4]
y= [0,2,4,6,8]
plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.',linestyle='--',markersize=10,markeredgecolor='blue')
plt.title('Basic Graph', fontdict={'fontname': 'Comic Sans MS','fontsize': 20})
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```

```python
x= [0,1,2,3,4]
y= [0,2,4,6,8]
# Short hand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b.--',label='2x')

## Line number two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,'r',label='x^2')


plt.title('Previous Graph', fontdict={'fontname': 'Comic Sans MS','fontsize': 20})
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```

```python
x= [0,1,2,3,4]
y= [0,2,4,6,8]

#Resizing graph
plt.figure(figsize=(5,3),dpi=150)

# Short hand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b.--',label='2x')

## Line number two
x2 = np.arange(0,4.5,0.5)
#Part of a graph as line
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
#Other part of graph as a dot
plt.plot(x2[5:],x2[5:]**2,'r--')

#Title
plt.title('My Graph', fontdict={'fontname': 'Comic Sans MS','fontsize': 20})

plt.xlabel('X axis')
plt.ylabel('Y axis')

#X,Y tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])


plt.legend()
#dpi is 300 is good when saving so graph has high resolution
plt.savefig('mygraph.png',dpi=300) 

plt.show()
```

### Load FIFA Data

#### Histogram

```python
fifa = pd.read_csv('fifa_data.csv')
fifa=fifa.drop(fifa.columns[0],axis=1)
fifa.head(5)
```

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall,bins=bins)

plt.xticks(bins)

plt.xlabel('Skill Level')
plt.ylabel('Number of Players')
plt.title('Distribution of Player Skills in FIFA 2018')
plt.show()
```

#### Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels=['Left','Right']
plt.title('Foot Preference of Players')
plt.pie([left,right],labels=labels,autopct='%.2f')

plt.show()
```

#### Pie Chart 2

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight<125].count()[0]
light_medium = fifa.loc[(fifa.Weight>=125)&(fifa.Weight<150)].count()[0]
medium = fifa.loc[(fifa.Weight>=150)&(fifa.Weight<175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight>=175)&(fifa.Weight<200)].count()[0]
heavy = fifa.loc[(fifa.Weight>=200)].count()[0]

weights=[light, light_medium, medium,medium_heavy,heavy]
```

```python
labels=['<125','125-150','150-175','175-200','>200']

plt.title('Weight distribution of FIFA players (in lbs)')
plt.style.use('ggplot')

explode=(.4,.2,0,0,.4)
plt.pie(weights,labels=labels,autopct='%.2f %%',pctdistance=0.8,explode=explode)

plt.show()
```

```python
plt.title('Weight distribution of FIFA players (in lbs)')
plt.style.use('default')

explode=(.4,.2,0,0,.4)
plt.pie(weights,labels=labels,autopct='%.2f %%',pctdistance=0.8,explode=explode)

plt.show()
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

plt.figure(figsize=(6,9))
labels=['FC Barcelona','Real Madrid','NE England Revolution']
plt.boxplot([barcelona,madrid,revs],labels=labels)

plt.title('FIFA Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```

```python
plt.figure(figsize=(6,9))
labels=['FC Barcelona','Real Madrid','NE England Revolution']
boxes = plt.boxplot([barcelona,madrid,revs],labels=labels,patch_artist=True,medianprops={'linewidth':2})


for box in boxes['boxes']:
    #set edge clor
    box.set(color='#4286f4',linewidth=2)
    #fill color
    box.set(facecolor='#a0a0a0')
    
plt.title('FIFA Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```
